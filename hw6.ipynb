```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.pyplot import figure
import scipy.stats as stat
```

```python
X = range(5, 1500, 5)
sigma = 1
gamma = 0.5
```

```python
def calc_a(n: int) -> float:
    X_2 = np.sum(np.random.normal(0, sigma, n) ** 2)
    q_1 = stat.chi2.ppf((1 + gamma) / 2, n)
    q_2 = stat.chi2.ppf((1 - gamma) / 2, n)
    return X_2 / q_2 - X_2 / q_1
```

```python
iterval_a = []
for x in X:
    iterval_a.append(np.mean([calc_a(x) for _ in range(300)]))
```

```python
figure(figsize=(18, 6), dpi=80)
plt.plot(X, iterval_a, linewidth=2, label='A')
plt.grid()
plt.legend()
plt.show()
```

```python
def calc_b(n: int) -> float:
    X_2 = n * np.mean(np.random.normal(0, sigma, n)) ** 2
    q_1 = stat.norm.ppf((3 + gamma) / 4) ** 2
    q_2 = stat.norm.ppf((3 - gamma) / 4) ** 2
    return X_2 / q_2 - X_2 / q_1
```

```python
iterval_b = []
for x in X:
    iterval_b.append(np.mean([calc_b(x) for _ in range(300)]))
```

```python
figure(figsize=(18, 6), dpi=80)
plt.plot(X, iterval_b, linewidth=2, label='B')
plt.grid()
plt.legend()
plt.show()
```
